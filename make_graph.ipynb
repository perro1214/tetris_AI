```python
with open('score.csv', 'r') as file:
    Y = []
    for i,line in enumerate(file):
        if i%5 == 0:
            Y.append(0)
        #print(line.strip().split(',')[1])
        Y[i//5] += float(line.strip().split(',')[1])
Y = list(map(lambda x: x/5, Y))
```

```python
X = []
for i in range(0, len(Y)):
    X.append(i)
```

```python
import matplotlib.pyplot as plt
import numpy as np

res3=np.polyfit(X, Y, 15)
y3 = np.poly1d(res3)(X)
plt.scatter(X, Y,label='original_scatter')
plt.plot(X, Y,label='original')
plt.plot(X, y3,label='kinzi')
plt.legend()
plt.tight_layout()
plt.show()
```
